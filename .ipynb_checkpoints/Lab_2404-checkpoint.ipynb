```python
df1 = pd.read_csv\
('scipy-2017-tutorial-pandas-master/data/concat_1.csv')
df2 = pd.read_csv\
('scipy-2017-tutorial-pandas-master/data/concat_2.csv')
df3 = pd.read_csv\
('scipy-2017-tutorial-pandas-master/data/concat_3.csv')
```

```python
dff1 = pd.DataFrame(\
                   {
                       'A':[1,2,3,4,5],
                       'B':[100,200,300,400,500]
                   })
dff1
```

```python
dff2 = pd.DataFrame(\
                   {
                       'A1':[1,3,5,7,9],
                       'B':[100,300,500,700,900]
                   })
dff2
```

```python
dff1.merge(dff2,left_on='A',right_on='A1')
```

```python
dff3 = pd.DataFrame(\
                   {
                       'A':[1,3,5,7,9],
                       'B':[100,300,500,700,900]
                   })
dff3
```

```python
dff1.merge(dff3,on='A')
```

```python
pd.merge(dff1,dff2,left_on='A',right_on='A1')
```
